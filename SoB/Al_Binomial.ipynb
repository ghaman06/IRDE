```python
import numpy as np
import lumos.brdf.library
import lumos.brdf.tools
import lumos.plot
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
import lumos.brdf 
B, C = lumos.brdf.tools.binomial_fit("aluminum_brdf.csv", n = 1, m = 2, l1 = -3, l2 = 3, d = 3.0)

print("B = ", np.array_str(B, precision = 3))
print("C = ", np.array_str(C, precision = 3))

Aluminum_brdf = lumos.brdf.library.BINOMIAL(B, C, d = 3.0, l1 = -3)
```

```python
fig, ax = plt.subplots()

phi_i, theta_i, phi_o, theta_o, brdf = lumos.brdf.tools.read_brdf("aluminum_brdf.csv")
ax.semilogy(phi_o, brdf, 'k.', alpha = 0.5, zorder = 2)
lumos.plot.BRDF_1D(ax, Aluminum_brdf, incident_angles = np.unique(phi_i))

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel("Outgoing Zenith Angle", fontsize = 16)
plt.ylabel(r"BRDF $(sr^{-1})$", fontsize = 16)
plt.title("Aluminum BRDF", fontsize = 20)
plt.legend(fontsize = 14)
plt.show()
```
